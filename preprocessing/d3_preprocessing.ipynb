```python
import pandas as pd
import time
import calendar
```

Preprocessing for Year to Year Efficiency

```python
df = pd.read_csv("cleaned_data_FULL.csv")
df.head()

df['Created Date'] = pd.to_datetime(df['Created Date'], format = '%m/%d/%Y %H:%M:%S %p')
df['Closed Date'] = pd.to_datetime(df['Closed Date'], format = '%m/%d/%Y %H:%M:%S %p')
df['Incident Zip'] = df['Incident Zip'].astype(int)

def to_hours(td):
    return round(td.seconds/float(3600),1)

df['Efficiency'] = (df['Closed Date'] - df['Created Date']).apply(lambda x: to_hours(x))
df['Month'] = df['Created Date'].apply(lambda x: calendar.month_name[x.month])

df['Year'] = df['Created Date'].apply(lambda x: x.year)

df_alt = df.copy()

df_alt = df_alt[(df_alt['Incident Zip'] >= 10000) & (df_alt['Incident Zip'] <= 11697)] 

df_d3 = df_alt.groupby(['Incident Zip','Year'])['Efficiency'].mean().reset_index()

df_d3.columns =  ['zip','year','efficiency']

df_d3

df_d3.to_csv("nyc_efficiency_full.csv")
```

Preprocessing for Hour to Hour

```python
df_hour = pd.read_csv("cleaned_data.csv")

df_hour.head()

df_hour['Date'] = pd.to_datetime(df_hour['Created Date'], format = '%m/%d/%Y %I:%M:%S %p')

df_hour['Hour'] = df_hour.Date.apply(lambda x: x.hour)

df_hour.Date[0].day

df_hour['Day'] = df_hour.Date.apply(lambda x: x.day)
df_hour['Month'] = df_hour.Date.apply(lambda x: x.month)
df_hour['Year'] = df_hour.Date.apply(lambda x: x.year)

grouped = df_hour.groupby(['Day','Month','Year','Hour','Incident Zip'])['Created Date'].count().reset_index().groupby(['Incident Zip','Hour'])['Created Date'].sum().reset_index()
grouped = grouped[(grouped['Incident Zip'] >= 10000) & (grouped['Incident Zip'] <= 11697)]

grouped.columns = ['zip','hour','count']

grouped.zip = grouped.zip.astype(int)

grouped.to_csv("hour_count.csv")
```
